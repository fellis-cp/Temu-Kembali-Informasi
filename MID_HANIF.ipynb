```python
#M.HANIF FAJAR ANGGARA - 21106050083

#panda get
import pandas as pd

# get ntltk
import nltk
nltk.download('all')
from nltk.corpus import stopwords
nltk.download('stopwords')

# preprocessing get
import re
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.stem import PorterStemmer




```

```python
# baca csv yang di upload
data = pd.read_csv('DS1.csv', encoding='latin-1')
data.head()
```

```python
# rename colom

data.columns = ['label', 'text']

data.head()
```

```python

# buat list text
text = list(data['text'])


# Init WordNetLemmatizer and PorterStemmer
lemmatizer = WordNetLemmatizer()
ps = PorterStemmer()

#Get English stopwords dari NLTK
stop_words = set(stopwords.words('english'))
# buat array kosong untuk menyimpan preprocesed text
corpus = []

# loop setiap text dalam kolom text
for text in data['text']:
   # cleaning
    r = re.sub('[^a-zA-Z]', ' ', text)
    r = r.lower()

    # Tokenize
    words = nltk.word_tokenize(r)

      # Stem using PorterStemmer
    words = [ps.stem(word) for word in words]
     # Remove stopwords

    words = [word for word in words if word not in stop_words]
     # Lemmatize setiap kata dengan WordNetLemmatizer
    words = [lemmatizer.lemmatize(word) for word in words]

   # join
    r = ' '.join(words)
    corpus.append(r)

```

```python
#assign corpus to data['text']
data['text'] = corpus
data['text']
data.head()
```

```python
stop_words = set(stopwords.words('english'))

def preprocess(document):
  words = nltk.word_tokenize(document.lower())
  return [word for word in words if word not in stop_words and word.isalpha()]

processed_documents = [preprocess(document) for document in corpus]
processed_documents
```

```python
sql_keywords = ["AND", "NOT", "OR"]
sql_queries = []

for doc_text in corpus:
    sql_query = []
    for word in doc_text.split(" "):
        if word in sql_keywords:
            sql_query.append(word)
            break
        else:
            sql_query.append("dictkey='%s'" % word)
    sql_queries.append(" ".join(sql_query))

for i, query in enumerate(sql_queries):
    print(f"Query SQL untuk dokumen ke - {i + 1}: {query}")
```
